# Data Preprocess

```python
from sklearn.preprocessing import StandardScaler
```

```python
def preprocess(df):
    df = df.fillna(method='ffill')
  
    temp_cols = ['T_Supply', 'T_Return']
    df['Timestamp'] = pd.to_datetime(df['Timestamp'],utc=True)
    df['year'] = pd.to_datetime(df['Timestamp']).dt.year
    df['month'] = pd.to_datetime(df['Timestamp']).dt.month
    df = df.drop(columns=['Timestamp'])
    X = data.drop('T_Supply', axis=1) 
    y = data['T_Supply']

    df[temp_cols] = StandardScaler().fit_transform(df[temp_cols])
    
    df['avg_temp'] = df[temp_cols].mean(axis=1)
    
    return X,y
```

```python
def get_new_data():
    
    df_new = pd.read_csv('HVAC_NE_EC_19-21.csv')
    X_new,y_new = preprocess(df_new) 
    return X_new,y_new
```

```python
def detect_anomalies(X, y_true, y_pred, threshold=2):
  
  # Calculate absolute error
  abs_error = abs(y_true - y_pred)
  
  # Find anomalies above threshold
  anomalies = X[abs_error > threshold]
  
  # Add anomaly flag column
  anomalies['anomaly'] = True
  
  print(f"Detected {len(anomalies)} anomalies")
  
  return anomalies
```

```python
def log_actual_vs_predicted(y_true, y_pred):
  
  # Log actuals vs predictions
  log_df = pd.DataFrame({'actual': y_true, 
                         'predicted': y_pred})
  
  # Calculate error metrics
  from sklearn.metrics import mean_absolute_error
  mae = mean_absolute_error(y_true, y_pred)
  
  print('Mean Absolute Error:', mae)
  
  # Visualize predictions
  import matplotlib.pyplot as plt
  plt.scatter(y_true, y_pred)
  plt.plot(y_true, y_true, color='red')
  plt.xlabel('Actual')
  plt.ylabel('Predicted')
  plt.show()

  # Log to CSV
  log_df.to_csv('model_monitoring_log.csv', index=False)
```

```python
def verify_anomalies(anomalies_df):

  print(f"Identified {len(anomalies_df)} potential anomalies")

  # Simulate manual verification 
  anomalies_df['anomaly_verified'] = np.random.choice([True, False], size=len(anomalies_df))
  
  # Subset to verified anomalies
  verified_df = anomalies_df[anomalies_df['anomaly_verified']]

  print(f"{len(verified_df)} anomalies verified by domain expert")

  return verified_df

```

```python
def label_verified_data(X, verified_df):

  # Add anomaly label column
  X['anomaly'] = 0

  for index, row in verified_df.iterrows():
    anomaly_index = row['index']
    X.loc[anomaly_index, 'anomaly'] = 1

  return X
```
